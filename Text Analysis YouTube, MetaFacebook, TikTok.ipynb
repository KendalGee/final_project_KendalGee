# Text Analysis 

## Lexical Density Analysis, Term Frequency Analysis, Term Frequency-Inverse Document Frequency Analysis

After completing the data collection process, three text analysis techniques were used to analyze the characterization and language used around the term "harassment" in the collected policy documents. The first technique was **lexical density analysis**, which measures the ratio of content words to function words in a text and is used to determine the richness and complexity of the text. This technique was used to analyze the complexity and transparency of the policy documents around harassment.

The second technique used was **term frequency analysis**, which involves counting the number of times words appear in a piece of text. This technique was used to count the number of words in each policy document to indicate its robustness, identify the most frequent terms within the policy documents on harassment (as well as how many times harassment was mentioned) to analyze policy priorities. 

The third technique used was **TF-IDF (Term Frequency-Inverse Document Frequency) analysis**, a statistical measure used to evaluate the importance of a word in a document in a collection of documents. This technique was used to determine what words were important to the policy documents as a whole (not just appeared frequently) to establish the importance of responses and establish the importance of the term harassment within the policy documents to explore if it was clearly defined.

### YouTube

```python
# Establishing lexical density

with open("CommunityGuidelines_YouTube.txt", "r", encoding='utf-8') as f:
    text = f.read()

text_tokens = nltk.word_tokenize(text)

text_lower = [t.lower() for t in text_tokens if t.isalnum()]
text_stops = [t for t in text_lower if t not in stops]

text_clean = [WordNetLemmatizer().lemmatize(t, get_wordnet_pos(t)) for t in text_stops]

text_clean_slice = text_clean[0:600]
ld_results = len(set(text_clean_slice)) / len(text_clean_slice)

print("Lexical density:", ld_results)
```

YouTube has the second highest score, with 53% content words and 47% function words. Meta/Facebook has the lowest score, with 49% content words and 51% function words. 

TikTok's higher score could indicate that the policy uses more complex/technical language. Therefore, it is difficult to make conclusive arguments about the characterization of harassment and harassment policies from lexical density analysis alone.

## Term Frequency-Inverse Document Frequency (TF-IDF)

```python
#The following script is adapted from (Krisel, 2023) https://github.com/intro-to-text-analysis-SIPA-S23/syllabus 
# and Stack Overflow public questions forum https://stackoverflow.com/questions/71213250/how-to-get-tf-idf-value-of-a-word-from-all-set-of-documents
```

### YouTube

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
#Open and read the file
with open("CommunityGuidelines_YouTube.txt", "r", encoding='utf-8') as f:
    text = f.read()
    
# Add custom stop words
stop_words = nltk.corpus.stopwords.words('english')
custom_stop_words = ['google', 'youtube', 'policy', 'community', 'guidelines' ]
stop_words.extend(custom_stop_words)

# Calculate the TF-IDF scores for the words in the document
tfidf_vectorizer = TfidfVectorizer(stop_words=stop_words)
tfidf = tfidf_vectorizer.fit_transform([text])

# Get the top 10 words by tf-idf score
top_words = []
feature_names = tfidf_vectorizer.get_feature_names_out()
for index in np.argsort(tfidf[0].toarray()[0])[::-1][:10]:
    top_words.append((feature_names[index], tfidf[0].toarray()[0][index]))

# Print the top 10 words and their tf-idf scores
for word, score in top_words:
    print(word, score)
```

```python
# Transform the document to a tf-idf matrix
tfidf_matrix = tfidf_vectorizer.transform([text])

# Get the index of the word "harassment" in the feature vocabulary
harassment_index = tfidf_vectorizer.vocabulary_['harassment']

# Get the tf-idf score for the word "harassment"
harassment_tfidf = tfidf_matrix[0, harassment_index]

print(harassment_tfidf)
```

## Findings 

Following the analuysis of term frequency, this analysis turned to TF-IDF to measure what the most important words were in each document outside of frequency.

Looking at the results, we can see that the top 10 words with the highest TF-IDF score in YouTube's policy are "report," "help," "creators," "channel," "content," "video," "learn," "videos," "comments," and "want." This suggests that YouTube's policy focuses on resources for creators and the reporting of content. The TF-IDF score for the word "harassment" is low, which suggests that the word is not emphasized as much in YouTube's policy.

In Facebook's policy, the top 10 words with the highest TF-IDF score are "content," "people," "products," "help," "use," "terms," "information," "us," "bullying," and "report." The TF-IDF score for the word "harassment" is higher than that of YouTube, indicating that Facebook places more emphasis on this issue.

In TikTok's policy, the top 10 words with the highest TF-IDF score are "information," "content," "user," "services," "use," "privacy," "data," "us," "policy," and "terms." This suggests that TikTok's policy focuses on information and user privacy, as well as providing services and setting terms and policies. The TF-IDF score for the word "harassment" is the lowest among the three platforms, indicating that TikTok does not place as much emphasis on this issue as Facebook and YouTube.

Finally, the importance of the word "report" in all three platforms demonstrates the emphasis on user reporting as a prevention and response plan. 

# Conclusion

In conclusion, this study analyzed the harassment policies of YouTube, Facebook/Meta, and TikTok using lexical density analysis, term frequency analysis, and term frequency-inverse document frequency analysis. The findings suggest that each platform has different priorities when it comes to harassment policies. Facebook/Meta has the most comprehensive policy on harassment, with a content on proactive measures such as removing content and providing resources to users. TikTok has a stronger focus on sexual harassment and compliance with law enforcement requests, while YouTube's policy focuses more on reporting and support to victims.

The analysis also revealed differences in the complexity of the language used across the policies, with TikTok having the highest lexical density score. However, this could indicate a more robust policy or indicate that the policy is more difficult to comprehend. Therefore, it is challenging to draw conclusions on how transparent/comprehensive the policies are. The TF-IDF analysis showed that the top words in each policy document reflect the platform's priorities in addressing harassment, with Facebook/Meta placing the greatest emphasis on the issue of harassment. However, the TF-IDF score also showed that all three platforms place emphasis on the user reporting content rather than detailing their own policy prevention and response plan. 

Finally, while the findings of this study shed some light on the differences in priorities and approaches to harassment policies among the three platforms, they do not indicate that any of the platforms have a clearly defined policy response to harassment. In fact, the results suggest that the policies of all three platforms may be characterized as vague, as they lack specific details on how the platforms plan to prevent and address harassment. Furthermore, these findings do not indicate the effectiveness of each policy and further analysis would be needed to make conclusive arguments about the quality of each platform's harassment policy. Nonetheless, this study's findings suggest that Facebook/Meta and TikTok have more comprehensive harassment policies than YouTube.
